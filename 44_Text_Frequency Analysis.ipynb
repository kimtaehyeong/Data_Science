```python
import pandas as pd
import nltk
from nltk.stem.porter import PorterStemmer
from nltk.tokenize import RegexpTokenizer
from nltk.corpus import stopwords 
```

```python
f = open("data/trump_speech.txt", 'r', encoding="utf-8")
lines = f.read()
f.close()
```

```python
# lines = lines[0]
print(lines)
```

```python
# 토큰화(tokenizing) 작업
# 토큰 : 자연어처리를 위해 긴 문자열을 분석을 위한 작은 단위로 나누는 문자열
tokenizer = RegexpTokenizer('[\w]+')
tokenizer
```

```python
stop_words = stopwords.words('english')
```

```python
words =  lines.lower()
tokens = tokenizer.tokenize(words)
print(tokens)
```

```python
stopped_tokens = [i for i in list((tokens)) if not i in stop_words]
print(stopped_tokens)
stopped_tokens2 = [i for i in stopped_tokens if len(i)>1]
print(stopped_tokens2)
```

```python
word_count = pd.Series(stopped_tokens2).value_counts().head(20)
word_count
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(20,10))
word_count.plot(kind='bar')
plt.plot()
```
